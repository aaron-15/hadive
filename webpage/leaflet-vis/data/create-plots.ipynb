```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('output_.txt', names=['cam_id', 'date', 'count'], skiprows=3)
data.head()
```

```python
for idx in data.cam_id.unique():
    fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(4,2))
    d = data[data['cam_id'] == idx]
    ax1.plot(range(len(d)), d['count'], color='k', alpha=0.6, )
    fig.savefig('./plots/{}.jpg'.format(idx))
```
